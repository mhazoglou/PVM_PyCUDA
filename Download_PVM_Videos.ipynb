```python
import requests as req
import bs4
import shutil
import os
```

```python
'''
goodies thanks to: https://stackoverflow.com/questions/16694907/how-to-download-large-file-in-python-with-requests-py
'''

def download_file(url, path=''):
    local_filename = path + url.split('/')[-1]
    r = req.get(url, stream=True)
    with open(local_filename, 'wb') as f:
        shutil.copyfileobj(r.raw, f)

    return local_filename
```

```python
webpage = 'http://pvm.braincorporation.net/'
r = req.get(webpage)
```

```python
html_doc = r.content
```

```python
r.raw
```

```python
soup = bs4.BeautifulSoup(html_doc, 'html.parser')
```

```python
print(soup.prettify())
```

```python
outpath = '/home/mhazoglou/Downloads/'
root = outpath + 'PVM_set/'
if root.rsplit('/')[-2] not in os.listdir(outpath):
    os.mkdir(root)

for link in soup.find_all('a'):
    link_ext = link.get('href')
    if 'PVM_set/' in link_ext:
        print(download_file(webpage + link_ext, path=root))
```
