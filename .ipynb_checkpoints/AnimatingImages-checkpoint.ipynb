```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib tk
import numpy as np
```

```python
fig = plt.figure(figsize=(12,9))

fig.add_subplot(221)


def f(x, y):
    return np.sin(x) + np.cos(y)

def gen_func():
    global x, y
    x += np.pi / 15.
    y += np.pi / 20.
    yield f(x, y)

x = np.linspace(0, 2 * np.pi, 10)
print x
y = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
print y

val = gen_func().next()
im = plt.imshow(val, animated=True, cmap='gray')
print(val.shape)

fig.add_subplot(222)

im2 = plt.imshow(val, animated=True)

def updatefig(val):
    im.set_array(val)
    im2.set_array(val)
    return im, im2

ani = animation.FuncAnimation(fig, updatefig, frames=gen_func, interval=16, blit=True)
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

def f(x, y):
    return np.sin(x) + np.cos(y)

x = np.linspace(0, 2 * np.pi, 120)
y = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
# ims is a list of lists, each row is a list of artists to draw in the
# current frame; here we are just animating one artist, the image, in
# each frame
ims = []
for i in range(60):
    x += np.pi / 15.
    y += np.pi / 20.
    im = ax.imshow(f(x, y), animated=True)
    
    im2 = ax2.imshow(f(x, y), animated=True)
    ims.append([im, im2])

ani = animation.ArtistAnimation(fig, ims, interval=16, blit=True,
                                repeat_delay=0)

# ani.save('dynamic_images.mp4', writer='ffmpeg',fps=60)

plt.show()
```
