```python
%%cython --annotate

import numpy as np
cimport numpy as np
cimport cython

@cython.boundscheck(False) # turn off bounds-checking for entire function
@cython.wraparound(False)  # turn off negative index wrapping for entire function
@cython.cdivision(True)   # c-division is best division
def conv2D(np.ndarray[np.double_t, ndim=3] in_arr,
           np.ndarray[np.double_t, ndim=4] filters,
           int conv_stride):

    cdef int in_height   = in_arr.shape[0]
    cdef int in_width    = in_arr.shape[1]
    cdef int in_channels = in_arr.shape[2]
    
    cdef int f_height   = filters.shape[0]
    cdef int f_width    = filters.shape[1]
    cdef int n_filters  = filters.shape[2]
    
    cdef int out_height = (in_height - f_height) / conv_stride + 1
    cdef int out_width  = (in_width  - f_width ) / conv_stride + 1
    cdef np.ndarray[np.double_t, ndim=3] out_arr = np.zeros([out_height,
                                                             out_width,
                                                             n_filters],
                                                            dtype=np.float64
                                                           )
    
    cdef int n, y, x, i, j, k
    cdef float val = 0
    
    for j in range(out_height):
        for i in range(out_width):
            for k in range(n_filters):
                val = 0.0
                
                for y in range(f_height):
                    for x in range(f_width):
                        for n in range(in_channels):
                            val += in_arr[y + j * conv_stride, x + i * conv_stride, n] * filters[y, x, k, n]
                
                out_arr[j, i, k] = val
    
    return out_arr
```

```python
from pycuda import driver, compiler, gpuarray, tools

import os
os.environ['CUDA_DEVICE'] = '0'
# -- initialize the device
import pycuda.autoinit
```

```python
img = np.array(plt.imread('/media/sdb/Mr._Krabs.jpg'), dtype=np.double)
img = img[:, :-1, :]
img_gpu = gpuarray.to_gpu(img)

# img = img[:-1, ...].copy()
img_height = np.int32(img.shape[0])
img_width = np.int32(img.shape[1])
n_colors = np.int32(img.shape[2])

edge_filter_x = np.array([[1, 0, -1],
                          [2, 0, -2],
                          [1, 0, -1]])
edge_filter_y = np.array([[1, 2, 1],
                          [0, 0, 0],
                          [-1, -2, -1]])

edge_filter = np.zeros([2, 3, 3, n_colors])
for i in range(n_colors):
    edge_filter[0, :, :, i] = edge_filter_x
    edge_filter[1, :, :, i] = edge_filter_y   

edge_filter_gpu = gpuarray.to_gpu(edge_filter)
n_filters = np.int32(edge_filter.shape[0])
spacial_extent = np.int32(edge_filter.shape[1])
conv_stride = np.int32(1)

spatial_extent = np.int32(2)
pool_stride = np.int32(2)
img_max_pool_gpu = gpuarray.zeros([(img_height - spatial_extent) // pool_stride + 1, 
                                   (img_width - spatial_extent) // pool_stride + 1,
                                   n_colors], dtype=np.float64)
img_avg_pool_gpu = gpuarray.zeros_like(img_max_pool_gpu)
mask_gpu = gpuarray.zeros_like(img_gpu)

upsample_gpu = gpuarray.zeros_like(img_gpu)
scale = np.int32(2)
avg_pool_height, avg_pool_width, avg_pool_channels = img_avg_pool_gpu.shape

filter_size = np.int32(spacial_extent ** 2 * n_filters * n_colors)

out_height = np.int32((img_height - spacial_extent) / conv_stride + 1) 
out_width  = np.int32((img_width  - spacial_extent) / conv_stride + 1)

edges_gpu = gpuarray.to_gpu(np.zeros([out_height, out_width, n_filters], dtype=np.float64))

ceil = lambda x: int(mt.ceil(x))

threadsize = (32, 32, 1)
gridsize = (ceil(out_width / threadsize[0]), ceil(out_height / threadsize[1]), ceil(n_filters / threadsize[2]))
gridsize_pool = (ceil(img_max_pool_gpu.shape[0] / threadsize[0]),
                 ceil(img_max_pool_gpu.shape[1] / threadsize[1]),
                 ceil(n_colors / threadsize[2]))
```

```python
%%cython --annotate

import numpy as np
cimport numpy as np
cimport cython

@cython.boundscheck(False) # turn off bounds-checking for entire function
@cython.wraparound(False)  # turn off negative index wrapping for entire function
@cython.cdivision(True)   
def conv2D(np.ndarray[np.double_t, ndim=3] in_arr,
           np.ndarray[np.double_t, ndim=4] filters,
           int conv_stride):

    cdef int in_height   = in_arr.shape[0]
    cdef int in_width    = in_arr.shape[1]
    cdef int in_channels = in_arr.shape[2]
    
    
    cdef int n_filters  = filters.shape[0]
    cdef int f_height   = filters.shape[1]
    cdef int f_width    = filters.shape[2]
    
    cdef int out_height = (in_height - f_height) / conv_stride + 1
    cdef int out_width  = (in_width  - f_width ) / conv_stride + 1
    cdef np.ndarray[np.double_t, ndim=3] out_arr = np.zeros([out_height,
                                                             out_width,
                                                             n_filters],
                                                            dtype=np.float64
                                                           )
    
    cdef int l, m, n, i, j, k
    cdef float val = 0
    
    for j in range(out_height):
        for i in range(out_width):
            for k in range(n_filters):
                val = 0.0
                
                for l in range(f_height):
                    for m in range(f_width):
                        for n in range(in_channels):
                            val += in_arr[l + j * conv_stride, m + i * conv_stride, n] * filters[k, l, m, n]
                
                out_arr[j, i, k] = val
    
    return out_arr
```

```python
# %%timeit -n 100
edges = conv2D(img, edge_filter, conv_stride)
```

```python
# you need to fix the order of the filter kernels you want it reversed for
# convolution but not reversed on the channels

kernel_code = """

#define CUDART_INF_D __longlong_as_double(0x7ff0000000000000)
#define CUDART_NINF_D __longlong_as_double(0xfff0000000000000)

__global__ void conv2dKernel(double *in_arr,
                             uint in_width,
                             uint in_height,
                             uint in_channels,
                             double *filters,
                             uint spacial_extent,
                             uint conv_stride,
                             uint n_filters,
                             double *out_arr)
{
  uint stride_x = blockDim.x * gridDim.x;
  uint stride_y = blockDim.y * gridDim.y;
  uint stride_z = blockDim.z * gridDim.z;
  
  uint out_width  = (in_width  - spacial_extent) / conv_stride + 1;
  uint out_height = (in_height - spacial_extent) / conv_stride + 1;
  
  extern __shared__ double s[];
  //__shared__ double in_s[{0}][{1}][{2}];
  
  uint L_filters = spacial_extent * spacial_extent * in_channels * n_filters;
  
  uint t = threadIdx.x + threadIdx.y * blockDim.x + threadIdx.z * blockDim.x * blockDim.y;
  uint grid_stride = stride_x * stride_y * stride_z;
  
  uint max_idx_filter = L_filters - 1;
  for (uint idx = t; idx < L_filters; idx += grid_stride)
      s[idx] = filters[max_idx_filter - idx];
  
  /*
  #pragma unroll
  for (uint j = threadIdx.y + blockIdx.y * blockDim.y;
       j < {0}; 
       j += stride_y)
  {
    for (uint i = threadIdx.x + blockIdx.x * blockDim.x; 
         i < {1}; 
         i += stride_x)
    {
      for (uint k = threadIdx.z + blockIdx.z * blockDim.z;
           k < {2};
           k += stride_z)
      {
        in_s[j][i][k] = in_arr[j * in_width * in_channels 
                              + i * in_channels + k];
      }
    }
  }
  */
  __syncthreads();
  
  double val;

  for (uint j = threadIdx.y + blockIdx.y * blockDim.y;
       j < out_height; 
       j += stride_y)
  {
    for (uint i = threadIdx.x + blockIdx.x * blockDim.x; 
         i < out_width; 
         i += stride_x)
    {
      for (uint k = threadIdx.z + blockIdx.z * blockDim.z;
           k < n_filters;
           k += stride_z)
      {
        val = 0.0;
        
        for (uint m = 0; m < spacial_extent; m += 1){
          for (uint l = 0; l < spacial_extent; l += 1){
            for (uint n = 0; n < in_channels; n += 1){
              val += s[k * in_channels * spacial_extent * spacial_extent 
                       + m * in_channels * spacial_extent + l * in_channels + n]
                     * in_arr[(m + j * conv_stride) * in_width * in_channels 
                              + (l + i * conv_stride) * in_channels + n];
            }
          }
        }
        
        out_arr[j * out_width * n_filters + i * n_filters + k] = val;
      }
    }
  }
}

__global__ void maxPoolKernel(double *in_arr,
                              uint in_width,
                              uint in_height,
                              uint in_channels,
                              uint spacial_extent,
                              uint pool_stride,
                              double *out_arr,
                              double *mask)
{
  uint stride_x = blockDim.x * gridDim.x;
  uint stride_y = blockDim.y * gridDim.y;
  uint stride_z = blockDim.z * gridDim.z;
  
  uint out_width  = (in_width  - spacial_extent) / pool_stride + 1;
  uint out_height = (in_height - spacial_extent) / pool_stride + 1;
  
  double val, temp_val;
  uint arg;
  
  for (uint j = threadIdx.y + blockIdx.y * blockDim.y;
       j < out_height; 
       j += stride_y)
  {
    for (uint i = threadIdx.x + blockIdx.x * blockDim.x; 
         i < out_width; 
         i += stride_x)
    {
      for (uint k = threadIdx.z + blockIdx.z * blockDim.z;
           k < in_channels;
           k += stride_z)
      {
        val = CUDART_NINF_D;
        arg = 0;
        
        for (uint m = 0; m < spacial_extent; m += 1){
          for (uint l = 0; l < spacial_extent; l += 1){
            
            temp_arg = (m + j * pool_stride) * in_width * in_channels +
                  (l + i * pool_stride) * in_channels + k;
            temp_val = in_arr[temp_arg];
            
            if (temp_val > val){
              val = temp_val;
              arg = temp_arg;
            }
            mask[temp_arg] = 0;
          }
        }
        mask[arg] = 1;
        
        out_arr[j * out_width * in_channels + i * in_channels + k] = val;
      }
    }
  }
}

__global__ void avgPoolKernel(double *in_arr,
                              uint in_width,
                              uint in_height,
                              uint in_channels,
                              uint spacial_extent,
                              uint pool_stride,
                              double *out_arr)
{
  uint stride_x = blockDim.x * gridDim.x;
  uint stride_y = blockDim.y * gridDim.y;
  uint stride_z = blockDim.z * gridDim.z;
  
  uint out_width  = (in_width  - spacial_extent) / pool_stride + 1;
  uint out_height = (in_height - spacial_extent) / pool_stride + 1;
  
  double val;
  uint arg;
  
  for (uint j = threadIdx.y + blockIdx.y * blockDim.y;
       j < out_height; 
       j += stride_y)
  {
    for (uint i = threadIdx.x + blockIdx.x * blockDim.x; 
         i < out_width; 
         i += stride_x)
    {
      for (uint k = threadIdx.z + blockIdx.z * blockDim.z;
           k < in_channels;
           k += stride_z)
      {
        val = 0;
        
        for (uint m = 0; m < spacial_extent; m += 1){
          for (uint l = 0; l < spacial_extent; l += 1){
            
            arg = (m + j * pool_stride) * in_width * in_channels +
                  (l + i * pool_stride) * in_channels + k;
            val += in_arr[arg];
          }
        }
        
        out_arr[j * out_width * in_channels + i * in_channels + k] = val / (spacial_extent * spacial_extent);
      }
    }
  }
}


__global__ void NNupSampleKernel(double *in_arr,
                                 uint in_width,
                                 uint in_height,
                                 uint in_channels,
                                 uint scale,
                                 double *out_arr)
{
  uint stride_x = blockDim.x * gridDim.x;
  uint stride_y = blockDim.y * gridDim.y;
  uint stride_z = blockDim.z * gridDim.z;
  
  uint out_width  = scale * in_width;
  uint out_height = scale * in_height;
  
  for (uint j = threadIdx.y + blockIdx.y * blockDim.y;
       j < out_height; 
       j += stride_y)
  {
    for (uint i = threadIdx.x + blockIdx.x * blockDim.x; 
         i < out_width; 
         i += stride_x)
    {
      for (uint k = threadIdx.z + blockIdx.z * blockDim.z;
           k < in_channels;
           k += stride_z)
      {
        out_arr[j * out_width * in_channels + i * in_channels + k] = 
            in_arr[(j / scale) * in_width * in_channels + 
                   (i / scale) * in_channels + 
                   k];
      }
    }
  }
}
"""
```

```python
mod = compiler.SourceModule(kernel_code)
conv2d = mod.get_function('conv2dKernel')
conv2d.prepare(['P', np.int32, np.int32, np.int32, 'P', np.int32, np.int32, np.int32, 'P'])

maxPool = mod.get_function('maxPoolKernel')
maxPool.prepare(['P', np.int32, np.int32, np.int32, np.int32, np.int32, 'P', 'P'])

avgPool = mod.get_function('avgPoolKernel')
avgPool.prepare(['P', np.int32, np.int32, np.int32, np.int32, np.int32, 'P'])

NNupSample = mod.get_function('NNupSampleKernel')
NNupSample.prepare(['P', np.int32, np.int32, np.int32, np.int32, 'P'])
```
