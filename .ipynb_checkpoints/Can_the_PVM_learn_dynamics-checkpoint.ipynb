```python
L = 20
N_div = 10001
T_max = 100
deltaT = T_max / (N_div - 1)
gamma = 0.02
blah = np.sqrt(1 - gamma ** 2)
T = np.linspace(0, T_max, N_div)
T = np.tile(np.array([T]).T, (1, 4))
training_data = {}
testing_data = {}
train_amp = list(0.25 * np.random.rand(L))
test_amp = list(0.25 * np.random.rand(L))
for _, shift in zip(train_amp, [0.25] * L):
    start = np.zeros_like(T)
    start[:, 0:2] += np.random.rand()
    start[:, 2:4] += np.random.rand()
    phase = np.arccos(blah)
    a = 1. / np.sin(phase)
    pos = (1 - np.exp(- gamma * (T - start * T_max)) * a * np.sin(blah * (T - start * T_max) + phase) )* np.heaviside(T - start * T_max, 0) / (1. - a) 
    pos[:, 1::2] = np.heaviside(T[:, 2:] - start[:, 0::2] * T_max, 0) 
    training_data['{}, {}'.format(_, shift)] = np.tile(pos, [1, 1])
    
for _, shift in zip(test_amp, [0.25] * L):
    start = np.zeros_like(T)
    start[:, 0:2] += np.random.rand()
    start[:, 2:4] += np.random.rand()
    phase = np.arccos(blah)
    a = 1. / np.sin(phase)
    pos = (1 - np.exp(- gamma * (T - start * T_max)) * a * np.sin(blah * (T - start * T_max) + phase) )* np.heaviside(T - start * T_max, 0) / (1. - a) 
    pos[:, 1::2] = np.heaviside(T[:, 2:] - start[:, 0::2] * T_max, 0)
    testing_data['{}, {}'.format(_, shift)] = np.tile(pos, [1, 1])
```
